```python
#Comenzamos con la prepracion de los datos (Data Cleaning)


#Primero renombramos las columnas para una mejor comprensiÃ³n tanto del analista de datos como para posibles consultores 

df_creditos.rename(columns={ 
    "ID": "ID del Cliente", 
    "Age": "Edad del Cliente",
    "Experience":"AÃ±os Experiencia", 
    "Income":"Ingresos Anual",  
    "ZIP Code":"Codigo Postal", 
    "Family":"Miembros Familia",
    "CCAvg":"Promedio Gasto TDC", 
    "Education": "Nivel de EducaciÃ³n",
    "Mortgage": "Valor de Hipoteca",
    "Personal Loan":"Cliente ACCA",
    "Securities Account": "Cliente con Cuenta", 
    "CD Account": "Cliente con Deposito", 
    "Online":"Cliente Digital",
    "CreditCard": "Cliente con TDC"}, inplace=True)

df_creditos
```

```python
# ExplicaciÃ³n de cada columna
'''
ID del Cliente: NÃºmero asignado al cliente
Edad del cliente: Edad del cliente en aÃ±os 
AÃ±os Experiencia: AÃ±os de experiencia laboral
Ingresos Anual: Ingresos anual en dolares 
Codigo Postal: Codigo postal de su residencia
Miembros Familia: NÃºmero de miembre de la familia
Promedio Gasto TDC: Gasto promedio mensual en tarjeta de credito en dolares
Nivel de EducaciÃ³n: Nivel de educacion del cliente (1: No graduado, 2: Graduado, 3: Profesional)
Valos de Hipoteca: Monto de la deuda en credito hipotecacio en dolares 
Cliente ACCA: Cliente aceptÃ³ credito en capaÃ±a anterior 
Cliente Cuenta: Cliente tiene una cuenta con activos financieros con el banco 
Cliente con Deposito: El cliente tine una cuenta certificada de deposito en el banco
Cliente Digital: Cliente usa medios por internet 
Cliente con TDC: El cliente tiene otra tarjeta de credito de otro banco

''';
```

```python
# Transformamos el contendio de cada columna para mayor comprensiÃ³n 
```

```python
#Multiplicamos por 1000 las columnas Ingresos Anual, Promedio Gasto TDC y Valor de Hipoteca  para ver su valor real 

df_creditos["Ingresos Anual"]= df_creditos["Ingresos Anual"]*1000
df_creditos["Promedio Gasto TDC"]= (df_creditos["Promedio Gasto TDC"]*1000).astype(int)
df_creditos["Valor de Hipoteca"]= (df_creditos["Valor de Hipoteca"]*1000).astype(int)

df_creditos
```

```python
# Tranformamos el contenido de las columnas: Nivel de EducaciÃ³n, Cliente ACCA, Cliente con Cuenta, Cliente con Deposito, Cliente Digital y Cliente con TDC; de nÃºmeros a verlos con su valor en letras para ser mas comprensivo. 
```

```python
mapeo_nivel_educativo = {
    1: "No graduado",
    2: "Graduado",
    3: "Profesional"
}

df_creditos["Nivel de EducaciÃ³n"] = df_creditos["Nivel de EducaciÃ³n"].map(mapeo_nivel_educativo)
```

```python
mapeo_cliente_acca = {
    0: "No",
    1: "Si"
}

df_creditos["Cliente ACCA"]=df_creditos["Cliente ACCA"].map(mapeo_cliente_acca)
```

```python
mapeo_cliente_cuenta = {
    0: "No",
    1: "Si"
}

df_creditos["Cliente con Cuenta"]= df_creditos["Cliente con Cuenta"].map(mapeo_cliente_cuenta)
```

```python
mapeo_cliente_deposito = {
    0: "No",
    1: "Si"
}

df_creditos["Cliente con Deposito"]= df_creditos["Cliente con Deposito"].map(mapeo_cliente_deposito)
```

```python
mapeo_cliente_digital = {
    0: "No",
    1: "Si"
}

df_creditos["Cliente Digital"]=df_creditos["Cliente Digital"].map(mapeo_cliente_digital)
```

```python
mapeo_cliente_TDC = {
    0:"No",
    1 : "Si"
}

df_creditos["Cliente con TDC"]=df_creditos["Cliente con TDC"].map(mapeo_cliente_TDC)
```

```python
# Creamos nuevas columnas que sean de interes para el analisis 
```

```python
# Creamos la columna Ingresos Mensual para ver el detalle de los ingresos por mes 

df_creditos["Ingresos Mensual"] = (df_creditos["Ingresos Anual"]/12).astype(int)
```

```python
# Creamos la columna Tipo de Cliente segmentandolos por los ingresos mensuales 

def clasificar_cliente(ingresos):
    if ingresos <= 1000:
        return "Clientes BÃ¡sicos"
    elif 1001 <= ingresos <= 4999:
        return "Banca Personal"
    elif 5000 <= ingresos <= 10000:
        return "Clientes Preferencial"
    elif ingresos > 10001:
        return "Clientes Oro"
    else:
        return "Sin Asignar"
    
df_creditos["Tipo de Cliente"]= df_creditos["Ingresos Mensual"].apply(clasificar_cliente)
```

```python
# Creamos la columna Cartera Edad para segmentar a los clientes segÃºn su Edad

def clasificar_edad(edad):
    if 18 <= edad <= 30:
        return "Cartera Joven"
    elif 31 <= edad <= 45:
        return "Adulto y Familias"
    elif 46 <= edad <= 60:
        return "Adultos Consolidados"
    elif edad > 60:
        return "Jubilados y Adultos Mayores"
    else:
        return "No Asignado"

df_creditos["Cartera Por Edad"] = df_creditos["Edad del Cliente"].apply(clasificar_edad)
```
